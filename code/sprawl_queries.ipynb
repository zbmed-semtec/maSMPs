```python
def generate_graph(file_path):
    """
    input: turtle file(.ttl)
    output: graph
    generating and serializing graph
    """
    graph = rdflib.Graph()
#     RDF_FILE = r'C:\Users\admin\Dhwani\Trial\GO.ttl'
    graph.parse(file_path, format='turtle')
    v = graph.serialize(format="ttl")
    return graph
```

```python
def get_allobjProperties(graph):
    query8=graph.query("""
    SELECT ?classID ?domain ?label ?type ?range ?union
    WHERE {
      ?classID rdfs:domain  ?domain .
      ?classID rdfs:label  ?label .
      ?classID rdf:type ?type .
      ?classID rdfs:range ?range .
      ?classID
      
      

      
   

      
      
         
     }""")
    pd.DataFrame(query8.bindings)
    # converts everything to strings including missing values
    pd.DataFrame(query8.bindings).applymap(str).rename(columns=str)

    # serialize with json and then parse (clobbers types, converting values to strings)
    import json
    results_json = query8.serialize(format="json")
    bindings = json.loads(results_json)["results"]["bindings"]
    bindings = [{k: v["value"] for k, v in result.items()} for result in bindings]
    df_all = pd.DataFrame(bindings)
    # print(bindings)
    return df_all
```

```python
def create_dataframe(file_path):
    """
    input: turtle file(.ttl)
    output: dataframe
    Returning dataframe created from graph
    """
    graph = generate_graph(file_path)
    df = to_dataframe(graph)
#     print(df.columns)
    return df
```
